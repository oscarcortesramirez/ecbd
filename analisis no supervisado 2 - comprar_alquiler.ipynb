```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Cargar los datos desde la muestra proporcionada
data = pd.read_csv("/Users/oscarcortes/Downloads/comprar_alquilar.csv")
```

```python
# Separar las características (X) y la variable objetivo (y)
X = data.drop('comprar', axis=1)
y = data['comprar']
```

```python
# Estandarizar los datos para que tengan media cero y desviación estándar uno
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Aplicar PCA para reducir la dimensionalidad a 2 componentes principales
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Crear un nuevo DataFrame con los componentes principales y la variable objetivo
pca_df = pd.DataFrame(data=X_pca, columns=['Componente 1', 'Componente 2'])
pca_df['comprar'] = y
```

```python
# Visualización de los resultados
plt.figure(figsize=(10, 6))
plt.scatter(pca_df[pca_df['comprar'] == 0]['Componente 1'], pca_df[pca_df['comprar'] == 0]['Componente 2'], label='No Comprar', alpha=0.7)
plt.scatter(pca_df[pca_df['comprar'] == 1]['Componente 1'], pca_df[pca_df['comprar'] == 1]['Componente 2'], label='Comprar', alpha=0.7)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('PCA - Reducción de Dimensionalidad')
plt.legend()
plt.show()
```

```python
# Variabilidad explicada por cada componente principal
explained_variance_ratio = pca.explained_variance_ratio_
```

```python
# Imprimir la variabilidad explicada por cada componente
print(f'Variabilidad explicada por el Componente 1: {explained_variance_ratio[0]:.2f}')
print(f'Variabilidad explicada por el Componente 2: {explained_variance_ratio[1]:.2f}')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Cargar los datos desde la muestra proporcionada
data = pd.read_csv('/Users/oscarcortes/Downloads/comprar_alquilar.csv')

# Separar las características (X) y la variable objetivo (y)
X = data.drop('comprar', axis=1)
y = data['comprar']

# Estandarizar los datos para que tengan media cero y desviación estándar uno
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Aplicar PCA para reducir la dimensionalidad a 2 componentes principales
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Crear un nuevo DataFrame con los componentes principales y la variable objetivo
pca_df = pd.DataFrame(data=X_pca, columns=['Componente 1', 'Componente 2'])
pca_df['comprar'] = y

# Visualización de los resultados
plt.figure(figsize=(10, 6))
plt.scatter(pca_df[pca_df['comprar'] == 0]['Componente 1'], pca_df[pca_df['comprar'] == 0]['Componente 2'], label='No Comprar', alpha=0.7)
plt.scatter(pca_df[pca_df['comprar'] == 1]['Componente 1'], pca_df[pca_df['comprar'] == 1]['Componente 2'], label='Comprar', alpha=0.7)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('PCA - Reducción de Dimensionalidad')
plt.legend()
plt.show()

# Variabilidad explicada por cada componente principal
explained_variance_ratio = pca.explained_variance_ratio_

# Imprimir la variabilidad explicada por cada componente
print(f'Variabilidad explicada por el Componente 1: {explained_variance_ratio[0]:.2f}')
print(f'Variabilidad explicada por el Componente 2: {explained_variance_ratio[1]:.2f}')
```
